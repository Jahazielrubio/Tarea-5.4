```python
import numpy as np
import matplotlib.pyplot as plt

# Datos del problema
x = np.array([5, 10, 15, 20, 25])  # Carga en kN
y = np.array([0.6, 1.2, 1.9, 2.5, 3.1])  # Elongaci√≥n en mm

# C√°lculo de los coeficientes de regresi√≥n lineal
n = len(x)
sum_x = np.sum(x)
sum_y = np.sum(y)
sum_xy = np.sum(x * y)
sum_x2 = np.sum(x**2)

# F√≥rmulas de regresi√≥n lineal
b = (n * sum_xy - sum_x * sum_y) / (n * sum_x2 - sum_x**2)
a = (sum_y - b * sum_x) / n

print(f"Coeficientes de la regresi√≥n:")
print(f"a (intercepto) = {a:.4f} mm")
print(f"b (pendiente) = {b:.4f} mm/kN")

# Predicci√≥n usando el modelo
y_pred = a + b * x

# Gr√°fica
plt.figure(figsize=(10, 6))
plt.plot(x, y, 'o', markersize=8, label='Datos experimentales')
plt.plot(x, y_pred, '-', linewidth=2,
         label=f'Ajuste lineal: y = {a:.3f} + {b:.3f}x', color='red')
plt.xlabel('Carga aplicada (kN)', fontsize=12)
plt.ylabel('Elongaci√≥n (mm)', fontsize=12)
plt.title('Relaci√≥n Carga-Elongaci√≥n en Barra de Acero', fontsize=14)
plt.legend(fontsize=11)
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(np.arange(0, 26, 5))
plt.yticks(np.arange(0, 3.5, 0.5))

# Mostrar ecuaci√≥n en el gr√°fico
plt.text(5, 2.8, f'y = {a:.3f} + {b:.3f}x', fontsize=12,
         bbox=dict(facecolor='white', alpha=0.8))

plt.savefig('regresion_lineal_materiales.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos del experimento de conducci√≥n t√©rmica
x = np.array([0, 2, 4, 6, 8])  # Posici√≥n en cm
y = np.array([100, 92, 85, 78, 71])  # Temperatura en ¬∞C

# C√°lculo de los coeficientes de regresi√≥n lineal
n = len(x)
sum_x = np.sum(x)
sum_y = np.sum(y)
sum_xy = np.sum(x * y)
sum_x2 = np.sum(x**2)

# F√≥rmulas de regresi√≥n lineal
b = (n * sum_xy - sum_x * sum_y) / (n * sum_x2 - sum_x**2)
a = (sum_y - b * sum_x) / n

print(f"Coeficientes de la regresi√≥n:")
print(f"a (intercepto) = {a:.4f} ¬∞C")
print(f"b (pendiente) = {b:.4f} ¬∞C/cm")

# Estimaci√≥n en x = 5 cm
x_estimado = 5
y_estimado = a + b * x_estimado
print(f"\nTemperatura estimada en x = {x_estimado} cm: {y_estimado:.2f} ¬∞C")

# Predicci√≥n para la gr√°fica
y_pred = a + b * x

# Gr√°fica
plt.figure(figsize=(10, 6))
plt.plot(x, y, 'o', markersize=8, label='Datos experimentales')
plt.plot(x, y_pred, '-', linewidth=2,
         label=f'Regresi√≥n lineal: T(x) = {a:.2f} + {b:.2f}x', color='red')
plt.plot([x_estimado], [y_estimado], 'gs', markersize=8,
         label=f'Estimaci√≥n en x={x_estimado} cm')

plt.xlabel('Posici√≥n a lo largo de la barra (cm)', fontsize=12)
plt.ylabel('Temperatura (¬∞C)', fontsize=12)
plt.title('Perfil de Temperatura en Barra Conductora', fontsize=14)
plt.legend(fontsize=11)
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(np.arange(0, 9, 1))
plt.yticks(np.arange(70, 105, 5))

# Mostrar ecuaci√≥n y punto estimado
plt.text(1, 75, f'T(x) = {a:.2f} + {b:.2f}x', fontsize=12,
         bbox=dict(facecolor='white', alpha=0.8))
plt.annotate(f'{y_estimado:.2f}¬∞C', (x_estimado, y_estimado),
             textcoords="offset points", xytext=(0,10), ha='center')

plt.savefig('transferencia_calor_regresion.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos del experimento de caudal
x = np.array([50, 70, 90, 110, 130])  # Presi√≥n en kPa
y = np.array([15, 21, 27, 33, 39])     # Caudal en L/min

# C√°lculo de los coeficientes de regresi√≥n lineal
n = len(x)
sum_x = np.sum(x)
sum_y = np.sum(y)
sum_xy = np.sum(x * y)
sum_x2 = np.sum(x**2)

# F√≥rmulas de regresi√≥n lineal
b = (n * sum_xy - sum_x * sum_y) / (n * sum_x2 - sum_x**2)
a = (sum_y - b * sum_x) / n

print(f"Coeficientes de la regresi√≥n:")
print(f"a (intercepto) = {a:.4f} L/min")
print(f"b (pendiente) = {b:.4f} L/(min¬∑kPa)")

# Predicci√≥n para x = 100 kPa
x_pred = 100
y_pred = a + b * x_pred
print(f"\nCaudal estimado a {x_pred} kPa: {y_pred:.2f} L/min")

# Generar puntos para la l√≠nea de regresi√≥n
y_line = a + b * x

# Gr√°fica
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='blue', label='Datos experimentales', s=80)
plt.plot(x, y_line, color='red',
         label=f'Regresi√≥n lineal: y = {a:.2f} + {b:.2f}x')
plt.scatter([x_pred], [y_pred], color='green', s=100,
            label=f'Predicci√≥n a {x_pred} kPa')

# Configuraci√≥n del gr√°fico
plt.xlabel('Presi√≥n (kPa)', fontsize=12)
plt.ylabel('Caudal (L/min)', fontsize=12)
plt.title('Relaci√≥n Presi√≥n-Caudal en Tuber√≠a', fontsize=14)
plt.legend(fontsize=11)
plt.grid(True, linestyle='--', alpha=0.7)

# Mostrar ecuaci√≥n y predicci√≥n
plt.text(55, 37, f'y = {a:.2f} + {b:.2f}x', fontsize=12,
         bbox=dict(facecolor='white', alpha=0.8))
plt.annotate(f'{y_pred:.2f} L/min', (x_pred, y_pred),
             textcoords="offset points", xytext=(0,10), ha='center')

plt.savefig('caudal_tuberia_regresion.png', dpi=300, bbox_inches='tight')
plt.show()
```
